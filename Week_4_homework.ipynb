```
import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from  sklearn.preprocessing import OrdinalEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_validate
from  sklearn.metrics import *
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn import tree
from sklearn import model_selection
import plotly.express as px
```

```
df_train = pd.read_csv("train.csv")
df_test = pd.read_csv("test.csv")
```

```
df_test.head()
```

```
df_train = df_train.drop("Unnamed: 0", axis=1)
df_test = df_test.drop("Unnamed: 0", axis=1)
```

```
df_train = df_train.drop("MarketingFlags", axis=1)
df_test = df_test.drop("MarketingFlags", axis=1)
```

```
df_train = df_train.drop("MarketingFlags_content", axis=1)
df_test = df_test.drop("MarketingFlags_content", axis=1)
```

```
df_train.head()
```

```
df_test.head()
```

```
categorical_columns = ["brand", "category", "name"]
```

```
for col in categorical_columns:
    df_train[col] = df_train[col].astype('category')
```

```
df_train = pd.get_dummies(df_train, drop_first=True)
```

```
for col in categorical_columns:
    df_test[col] = df_test[col].astype('category')
```

```
df_test = pd.get_dummies(df_test, drop_first=True)
```

```
df_train.head()
```

```
df_test.head()
```

```
col_list = list(set().union(df_test.columns, df_train.columns))
```

```
len(col_list)
```

```
def add_missing_dummy_columns( d, columns ):
    missing_cols = set( columns ) - set( d.columns )
    for c in missing_cols:
        d[c] = 0
```

```
add_missing_dummy_columns(df_test, col_list)
```

```
add_missing_dummy_columns(df_train, col_list)
```

```
df_test.drop("rating", axis=1)
```

```
X = df_train.loc[:, df_train.columns != 'rating']
```

```
y = df_train["rating"]
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```
model=DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```
y_pred = model.predict(X_test)
```

```
accuracy_score(y_test, y_pred)
```

```
precision_score(y_test, y_pred)
```

```
recall_score(y_test, y_pred)
```

```
f1_score(y_test, y_pred)
```

```
df_test = df_test.drop("rating", axis=1)
```

```
y_pred_test = model.predict(df_test)
```

```
df_test["rating"] = y_pred_test
```
